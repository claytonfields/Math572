Use the EM algorithm to fit a $(1-\pi)$ : $\pi$ mixture of two Poission distributions, Poisson$(\lambda_{1})$ and Poisson$(\lambda_{2})$, to the displayed data.

### 1(a):
Derive the EM algorithm for the maximum likelihood estimates of $\pi, \lambda_{1}, \lambda_{2}$. Present your derivation in a Markdown cell in the Jupyter Notebook.

Let X = observed data, Z = missing data, Y = complete data

$f(x; \Pi, \lambda_1,\lambda_2) = (1-\Pi_1)*Poi(\lambda_1) + \Pi*Poi(\lambda_2)$

Our missing data is $Z = (z_1, ..., z_n)^T $, where $z_j$ represents which component of the distribution the jth data comes from the first($z=0$) or second($z=1$).

Let $\theta = (\Pi, \lambda_1, \lambda_2)  $

$L(\theta; y) = [\prod_{i=1}^n [(1-z_i)*Poi(\lambda_1) + \Pi*Poi(\lambda_2)][\Pi^{z_i}(1-\Pi)^{1-z_i}] $ 

$l(\theta; y) = \sum_{i=1}^n (1-z_i[Poi(\lambda_1) + z_iPoi(\lambda_1)] + \sum_{i=1}^n[z_i\log(1-\Pi) + z_i*\log(1-z_i)]$

Then we have 

$Q(\theta|)\theta^{(t)} = \sum_{i=1}^n [(1-z_i^{(t)}Poi(\lambda_1) + z_i^{(t)}Poi(\lambda_1)] + \sum_{i=1}^n[z_i^{(t)}\log(\Pi) + (1-z_i^{(t)})*\log(1-\Pi)]\big)$

Where $z_i^{(t)} = E\big(z_i|\theta)\frac{\Pi^{(t)}Poi(\lambda_2}{f(x;\theta)}$

$\Pi: \frac{dQ}{d\Pi} = 0 = \sum_{i=1}^n \frac{1-z_i^{(t)}}{1-\Pi} + \frac{z_i^{(t)}}{\Pi}$ 

$\Pi = \sum_{i=1}^n \frac{z_i^{(t)}}{n}$

$\lambda_1:  \frac{dQ}{d\lambda_1} = \sum_{i=1}^n (1-z^{(t)})(\frac{x_i}{\lambda_1}-1) = 0$

$ \frac{\sum_{i=1}^n(x_i - x_iz_i)}{\lambda_1} - \sum_{i=1}^n(1-z_i^{(t)}) = 0$

$\lambda_1 = \frac{\sum_{i=1}^n(x_i - x_iz_i^{(t)})}{\sum_{i=1}^n(1-z_i^{(t)}) }$

$\lambda_2: \frac{dQ}{d\lambda_2} = \sum_{i=1}^n z_i^{(t)}(\frac{x_i}{\lambda_2}-1) = 0$

$\frac{\sum_{i=1}^n x_iz_i^{(t)}}{\lambda_2} = \sum z_i^{(t)}$

$\lambda_2 = \frac{\sum_{i=1}^n x_iz_i^{(t)}}{\sum_{i=1}^n z_i^{(t)}}$

Implement the EM algorithm for this data set. Present the output of your code in the format similar to Table 4.1 on page 102 of the book Computational Statistics.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.stats import poisson, relfreq, multivariate_normal
from scipy.linalg import norm

x = np.vstack((np.zeros((162,1)),np.ones((267,1)),2*np.ones((271,1)),
               3*np.ones((185,1)), 4*np.ones((111,1)),
               5*np.ones((61,1)), 6*np.ones((27,1)), 7*np.ones((8,1)),
               8*np.ones((3,1)), 9*np.ones((1,1))))

def pihat(z):
    n = z.size
    return np.sum(z/n)

def l1hat(x,z):
    return np.sum(x-z*x)/np.sum(1-z)

def l2hat(x,z):
    return np.sum(z*x)/np.sum(z)

def f(x,pi,l1,l2):
    return (1 - pi)*poisson.pmf(x,l1)+pi*poisson.pmf(x,l2)

def E(x,pi,l1,l2):
    a = pi*poisson.pmf(x,l2)
    b = (1 - pi)*poisson.pmf(x,l1)+pi*poisson.pmf(x,l2)
    return a/b

maxiter = 1000
pi = .359
l1 = 2.5
l2 = 1.2

pil = [pi]
l1l = [l1]
l2l = [l2]
ls  = []
theta = []

for i in range(maxiter):
#     print("pi: %f, l1: %f, l2: %f"%(pi, l1, l2))
    z = E(x,pi,l1,l2)
    pi = pihat(z)
    l1 = l1hat(x,z)
    l2 = l2hat(x,z)
    pil.append(pi)
    l1l.append(l1)
    l2l.append(l2)
    theta.append([pi,l1,l2])
    ls.append([i,pi,l1,l2])

Dpi = []
Dl1 = []
Dl2 = []
R = []
for i in range(999):
    ls[i].append(norm(np.array(theta[i+1])-np.array(theta[i]))/norm(np.array(theta[i])))
    ls[i].append(np.round((pil[i+1] - pil[-1])/(pil[i]-pil[-1]),3))
    ls[i].append(np.round((l1l[i+1] - l1l[-1])/(l1l[i]-l1l[-1]),3))
    ls[i].append(np.round((l2l[i+1] - l2l[-1])/(l2l[i]-l2l[-1]),3))
   
    #ls[i].append(R[i],Dpi[i],Dl1[i],Dl2[i])
# print(Dpi)
# print(Dl1)
# print(Dl2)



df = pd.DataFrame(ls[970:980], columns = ['i','pi','lambda1','lambda2',"R","D_pi","D_L1","D_L2"])
df = df.style.hide_index()
df = df.set_table_styles([dict(selector='th', props=[('text-align', 'center')])])
df
```

## Question 2:
Suppose that the probability density function (PDF) of a bivariate random vector $X = (X_{1},X_{2})^{T}$ is a mixture of bivariate normal

$f(x;\theta) = (1-\pi)\phi(x;\mu_{1},\Sigma_{1})+\pi\phi(x;\mu_{2},\Sigma_{2}),$

where $\phi(x;\mu,\Sigma)$ is the bivariate normal PDF with mean $\mu$ and covariance matrix $\Sigma$, and $\theta = [\pi,\mu_{1},\mu_{2},\Sigma_{1},\Sigma_{2}].$

The data set data_mvnorm2mix.csv is a random sample from these two normal mixtures, implement the EM algorithm to find the maximum likelihood estimate (MLE) of the parameter $\theta$. Present your result as following:


### 2(a):
Print the MLE of the parameter $\theta$.

```python
"""
Problem 2
"""

    
data = pd.read_csv('data_mvnorm2mix.csv',sep=',',header=None).to_numpy()


def pihat(z):
    n = z.size
    return np.sum(z)/n

def mu1hat(z,x):
    n = z.size
    l = []
    for i in range(n):
        l.append(((1-z[i])*x[i].T))
    num = np.sum(l,axis=0).reshape(2,1)
    den = (n-np.sum(z))
    return num/den

def mu2hat(z,x):
    n = z.size
    l = []
    for i in  range(n):
        l.append((z[i]*x[i].T))
    num = (np.sum(l,axis=0).reshape(2,1))
    den = np.sum(z)
    return num/den

def sig1hat(z,x,mu1):
    n= z.size
    l = []
    for i in range(n):
        l.append(((1-z[i])*(x[i].reshape(2,1)-mu1.reshape(2,1))@(x[i].reshape(2,1)-mu1).T))
    num = np.sum(l,axis=0)
    den = (n-np.sum(z))
    return num/den

def sig2hat(z,x,mu2):
    n = z.size
    l = []
    for i in range(n):
        l.append((z[i]*(x[i].reshape(2,1)-mu2)@(x[i].reshape(2,1)-mu2).T))
    num = np.array(l).sum(axis=0)
    den = np.sum(z)
    return num/den



def fx(x,pi,mu1,mu2,sig1,sig2):
    term1 = (1-pi)*multivariate_normal.pdf(x,mean = mu1.reshape(2,),cov=sig1)
    term2 =  pi*multivariate_normal.pdf(x,mean = mu2.reshape(-1,),cov=sig2)
    return term1 + term2
       
        
def Ez(x,pi,mu1,mu2,sig1,sig2):
    return pi*multivariate_normal.pdf(x,mean = mu2.reshape(-1,),cov=sig2)/fx(x,pi,mu1,mu2,sig1,sig2)

def L(data,pi,mu1,mu2,sig1,sig2):
    n = z.size
    l = []
    for i in range(n):
        term1 = (1-z[i])*np.log(multivariate_normal.pdf(data[i],mean = mu1.reshape(-1,),cov=sig1))
        term2 = z[i]*np.log(multivariate_normal.pdf(data[i],mean = mu2.reshape(-1,),cov=sig2))
        term3 = (1-z[i])*np.log(1-pi)+z[i]*np.log(pi)
        l.append(term1+term2+term3)
    return np.sum(l)
        
maxiter = 25
pi = .6
mu1 = np.array([[1],[2]])
mu2 = np.array([[2],[4]])
sig1 = np.array([[2,1],[3,8]])
sig2 =  np.array([[1,-1],[-1,6]])
loglist = []

for i in range(maxiter):
    z = Ez(data,pi,mu1,mu2,sig1,sig2)
    z = z.reshape(800,1)
    pi = pihat(z)
    mu1 = mu1hat(z,data)
    mu2 = mu2hat(z,data)
    sig1 = sig1hat(z,data,mu1)
    sig2 = sig2hat(z,data,mu2)
    loglist.append(L(data,pi,mu1,mu2,sig1,sig2))
print("pi: ",pi)
print("mu1: ", mu1)
print("mu2: ", mu2)
print("sig1: ", sig1)
print("sig2: ", sig2)
```

```python
#Part C
def dis1(x,mu1,sig1):
    return multivariate_normal.pdf(x,mean = mu1.reshape(2,),cov=sig1)

def dis2(x,mu2,sig2):
    return multivariate_normal.pdf(x,mean = mu2.reshape(-1,),cov=sig2)

x1d1list = []
x1d2list = []
x2d1list = []
x2d2list = []

for i in range(800):
    if dis1(data[i],mu1,sig1)>dis2(data[i],mu2,sig2):
        x1d1list.append(data[i,0])
        x2d1list.append(data[i,1])
    else:
        x1d2list.append(data[i,0])
        x2d2list.append(data[i,1])
plt.figure()
plt.scatter(x1d1list,x2d1list)
plt.scatter(x1d2list,x2d2list)
plt.xlabel("x1")
plt.ylabel("x2")
plt.legend(("Distribution 1","Distribution 2"))
plt.title("Observation Classification")
```
